```python
import pandas as pd
import json
```

```python
DATASET_PATH = '../datasets/downsampled_50_50.csv'
EXPORTED_GENERATIONS_PATH = '../datasets/baseline_zero_shot_gpt4_50_50.csv'
PROMPT_NAME = 'prompt1_en_short_temp_0_gpt-4-0613'
```

```python
# Replace 'your_file.csv' with the actual CSV file path
df = pd.read_csv(EXPORTED_GENERATIONS_PATH)
```

```python
# remove column "metadata"
df = df.drop(columns=['metadata'])
df = df.drop(columns=['name'])


df["completion"] = df["completion"].apply(lambda x: x.replace('{"completion":', "").replace('}', "").replace('"',""))
try: 
    df["prompt"] = df["prompt"].apply(lambda x: json.loads(x)[0][0]["content"]) 
except:
    pass
```

```python
df_dataset = pd.read_csv(DATASET_PATH)
```

```python
# add column "label" and "predicted_label"
df['predicted_label'] = ""
df['label'] = ""
df['error'] = ""

# copy column "label" from df_dataset to df
for i in range(len(df)):
    label = df_dataset['label'][len(df_dataset)-i-1]
    df['label'][i] = label
    predicted_label = df_dataset[PROMPT_NAME][len(df_dataset)-i-1]
    df['predicted_label'][i] = predicted_label
    if label != predicted_label:
        if label == False:
            df['error'][i] = "FP"
        elif label == True:
            df['error'][i] = "FN"
        else:
            print("ERROR: " + label)
    else:
        df['error'][i] = "NONE"

```

```python
# Replace 'merged_file.csv' with the desired output file name
df.to_csv(EXPORTED_GENERATIONS_PATH.replace(".csv","") + '_merged.csv', index=True)
```
