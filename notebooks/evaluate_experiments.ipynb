```python
import pandas as pd
INPUT_PATH = '../datasets/downsampled_50_50.csv'
OUTPUT_PATH = '../evals/eval_downsampled_50_50_v1.csv'

df_data = pd.read_csv(INPUT_PATH, index_col=False)
df_eval = pd.read_csv(OUTPUT_PATH, index_col=False)

PROMPT_NAME = "prompt1_cot_manual_temp_0_gpt3.5-0301"
EVAL_COLUMN_NAME = "prompt1_cot_manual_temp_0_gpt3.5-0301_maxtok-15"

DATASET_NAME = "downsampled_50_50"

```

```python
df_data.head(10)
```

```python
from sklearn import metrics


def evalColumn(model, prompt, df, column):
    y_true = df["label"]
    y_pred = df[column]
    conf_matrix = metrics.confusion_matrix(y_true, y_pred)
    precision = metrics.precision_score(y_true, y_pred)
    recall = metrics.recall_score(y_true, y_pred)
    f1 = metrics.f1_score(y_true, y_pred)
    accuracy = metrics.accuracy_score(y_true, y_pred)
    global df_eval
    df_eval = df_eval._append({'Model': model, 'Prompt': prompt, 'Accuracy': accuracy, 'precision': precision, 'recall': recall, 'F1': f1, "confusion_matrix": conf_matrix, "Dataset":DATASET_NAME}, ignore_index=True)


evalColumn("openAI - gpt 3.5", EVAL_COLUMN_NAME, df_data, PROMPT_NAME)
    
```
